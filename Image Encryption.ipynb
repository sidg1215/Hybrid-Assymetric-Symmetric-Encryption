```python
from functools import reduce
def combineRows(blocks):
    rows = np.array([combineColumns(blocks[i]) for i in range(0,len(blocks))])
    return reduce((lambda a,b : np.concatenate((a,b))), rows)
    
def combineColumns(row):
    return reduce((lambda a,b : np.concatenate((a,b), axis = 1)), row)
```

```python
combineRows(blocks)
```

```python
combinedBlocks = combineRows(blocks)
combinedBlocks
```

```python
def makeRGBArray(arr):
    return np.array([arr[i].reshape(len(arr[i])//3, 3) for i in range(0, len(arr))])
```

```python
newImage = Image.fromarray(makeRGBArray(combinedBlocks), 'RGB')
```
